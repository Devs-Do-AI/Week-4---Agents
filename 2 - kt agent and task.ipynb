```python
from youtube_transcript_api.formatters import TextFormatter
from youtube_transcript_api import YouTubeTranscriptApi

# Must be a single transcript.
transcript = YouTubeTranscriptApi.get_transcript("ghDaTYbmC9o")

formatter = TextFormatter()

# .format_transcript(transcript) turns the transcript into a JSON string.
formatted = formatter.format_transcript(transcript)

# Now we can write it out to a file.
with open("2_kt_transcript.txt", "w", encoding="utf-8") as transcript_file:
    transcript_file.write(formatted)
```

```python
from crewai import Agent, Crew
from crewai_tools import FileReadTool

recap_writer = Agent(
    role="Kill Tony Recap Writer",
    goal="Write an engaging recap of the latest Kill Tony episode",
    backstory="""You are a comedy journalist who writes detailed, 
    humorous recaps of Kill Tony episodes for the show.
    Your recaps capture the best moments and inside jokes from each episode.""",
    tools=[
        FileReadTool(),  # Tool to read the transcript file
    ],
)
```

```python
from crewai import Task

recap_task = Task(
    description="Write a recap for the latest Kill Tony episode based on the transcript {transcript_file}",
    expected_output="A draft of a blog post with a detailed, engaging recap of the latest episode",
    agent=recap_writer,
)
```

```python
crew = Crew(agents=[recap_writer], tasks=[recap_task])
```

```python
result = crew.kickoff(
    inputs={
        "transcript_file": "2_kt_transcript.txt",  # The path to the transcript file
    }
)
```
